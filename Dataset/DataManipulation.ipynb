```python
import pandas as pd

# Create a list of the categories
categories = sorted_list

# Create a dictionary to map the categories to integers
cat_to_int = {cat: i for i, cat in enumerate(categories)}

# Create a function to one-hot encode a single category
def one_hot_encode(category):
    encoding = [0] * len(categories)
    encoding[cat_to_int[category]] = 1
    return encoding

# Example usage
category = "Python"
encoding = one_hot_encode(category)
print(encoding)

# Create a DataFrame of example data
data = pd.DataFrame({
    "Tool": sorted_list,
    "Category": sorted_list
})

# Apply one-hot encoding to the "Category" column
one_hot = pd.get_dummies(data["Category"], prefix="Category")

# Add the one-hot encoded columns to the DataFrame
data = pd.concat([data, one_hot], axis=1)
print(data.head())
```
